```python
import pandas as pd
```

```python
data1=pd.read_csv('desktop/data.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
data1.plot(x='CRIM',y='ZN',kind='hist', bins=70)
plt.xlabel('proportion of residential land zoned for lots over 25,000 sq.ft.')
plt.show()
data1.plot(x='CRIM',y='INDUS',kind='hist', bins=70)
plt.xlabel('proportion of non-retail business acres per town')
plt.show()
data1.plot(x='CRIM',y='CHAS',kind='hist', bins=70)
plt.xlabel('Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)')
plt.show()
data1.plot(x='CRIM',y='NOX',kind='hist', bins=70)
plt.xlabel('nitric oxides concentration (parts per 10 million)')
plt.show()
data1.plot(x='CRIM',y='RM',kind='hist', bins=70)
plt.xlabel('average number of rooms per dwelling')
plt.show()
data1.plot(x='CRIM',y='AGE',kind='hist', bins=70)
plt.xlabel('proportion of owner-occupied units built prior to 1940')
plt.show()
data1.plot(x='CRIM',y='DIS',kind='hist', bins=70)
plt.xlabel('weighted distances to five Boston employment centres')
plt.show()
data1.plot(x='CRIM',y='RAD',kind='hist', bins=70)
plt.xlabel('index of accessibility to radial highways')
plt.show()
data1.plot(x='CRIM',y='TAX',kind='hist', bins=70)
plt.xlabel('full-value property-tax rate per $10,000')
plt.show()
data1.plot(x='CRIM',y='PTRAT',kind='hist', bins=70)
plt.xlabel('pupil-teacher ratio by town')
plt.show()
data1.plot(x='CRIM',y='B',kind='hist', bins=70)
plt.xlabel(' 1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town')
plt.show()
data1.plot(x='CRIM',y='LSTAT',kind='hist', bins=70)
plt.xlabel(' % lower status of the population')
plt.show()
data1.plot(x='CRIM',y='MEDV',kind='hist', bins=70)
plt.xlabel('Median value of owner-occupied homes in $1000s')
plt.show()
```
